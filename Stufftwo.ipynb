```python
lenlist = []
aax = int(input('how many words? '))
for i in range(aax):
    x = input('enter a word: ')
    ax = len(x)
    lenlist.append(ax)
print(lenlist)
```

```python
lenlisttwo = []
aay = int(input('how many words? '))
for i in range(aay):
    y = input('enter a word: ')
    ay = len(y)
    if ay > 3:
        lenlisttwo.append(y)
print(lenlisttwo)
```

```python
lenlistthree = []
aaz = int(input('how many words? '))
for i in range(aaz):
    z = input('enter a word: ')
    az = z [::-1]
    lenlistthree.append(az)
print(lenlistthree)
```

```python
lst = []
def count_even_odd(lst):
    even = 0
    odd = 0
    for item in lst:
        if item %2 == 0:
            even += 1
        else:
            odd += 1
    finaldict = {'even':even, 'odd':odd}
    print(finaldict)
adarav = int(input('how many numbers? '))
for i in range(adarav):
    odorev = int(input('enter a number: '))
    lst.append(odorev)
count_even_odd(lst)
```

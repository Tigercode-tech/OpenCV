Recursion

```python
def fitness():
    z = 0
    n = int(input('enter the natural number'))
    for j in range(1,n+1):
        z += j
    print(z)
fitness()
    
```

```python
def fitness_recursion(n):
    if n == 0 or n == 1:
        return(n)
    else:
        return fitness_recursion(n-1)+n

n = int(input('enter a number.'))
fitness_recursion(n)
```

```python
def fact(m):
    if m == 0 or m == 1:
        return(m)
    else:
        return fact(m-1)*m

m = int(input('enter a number.'))
fact(m)
    
```

```python
def cube(x, y):
    if x == 1:
        return x
    elif y == 0:
        return 1
    else:
        return cube(x, y-1)*x
cube(3, 4)
```

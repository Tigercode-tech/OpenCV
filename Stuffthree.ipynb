```python

def count(z):
    y = 0
    for i in range (1, z+1):
        y = y + i
    print(y)
count(int(input('How many numbers?')))
```

```python
x = int(input("enter a number"))
for a in range(2,x):
    if x%a == 0:
        print("the number is not prime")
        break
else:
        print('the number is prime')
```

```python
stri = input('Enter a sentence')
vount = 0
vlist = ['a', 'e', 'i', 'o', 'u', 'y']
for letter in stri:
    if letter in vlist:
        vount += 1
print("there are", vount, "vowels in your sentence")
```

```python
def am():
    ex = 0
    fin = 0
    k = int(input("enter a number"))
    ex = len(str(k))
    for numbert in str(k):
        fin += int(numbert)**ex
    if fin == k:
        print('armstrong number')
    else:
        print(':(')
am()
```

```python
def am():
    ex = 0
    fin = 0
    k = int(input("enter a number"))
    ex = len(str(k))
    for numbert in str(k):
        fin += int(numbert)**ex
    if fin == k:
        print('armstrong number')
    else:
        print(':(')
am()
```

```python
numb = int(input('Enter an\ number'))
numb2 = numb
bmun = 0
while numb > 0:
    dig = numb%10
    bmun = (bmun * 10) + dig
    numb = numb//10
if numb2 == bmun:
    print('Palindrome! :D')
else:
    print(':(')
```

```python
try:
    x = int(input('Enter a number'))
    y = int(input('Enter a number'))
    print(x/y)
except(ValueError):
    print('The code seems to have run in to an error, please enter a number next time.')
except(ZeroDivisionError):
    print('The code seems to have run in to an error, please do not enter zero next time.')
except:
    print('The code seems to have run in to an error, please do not use this program next time.')
```
